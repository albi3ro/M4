```julia
# importing our packages
using Gadfly;
```

Define The Relevant Variables

```julia
lattice="triangular"
if(lattice=="square")
    d=2;
    a=[1,0,0];
    b=[0,1,0];
    c=[0,0,1]
elseif(lattice=="triangular")
    d=2;
    a=[1,0];
    b=[.5,sqrt(3)/2];
    c=[0,0,1];
end

Nx=3;
Ny=3;
Nz=1;
```

```julia
N=Nx*Ny
aM=transpose(a);
bM=transpose(repeat(b,outer=[1,Nx]));

X=Array{Float64}(N,2);
```

```julia
for i in 1:Nx
    X[i,:]=(i-1)*a;
end

for j in 2:Ny
    X[(Nx*(j-1)+1):(Nx*(j-1)+Nx),:]=X[1:Nx,:]+(j-1)*bM;
end
```

```julia
plot(x=X[:,1],y=X[:,2])
```
